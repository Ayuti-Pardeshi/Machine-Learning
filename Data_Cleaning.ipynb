```python
# header files
import pandas as pd
```

```python
data = pd.read_excel('Ayuti.xlsx')
# reading the file using in-built function read_excel
```

```python
df = data.drop_duplicates()
```

```python
df
```

```python
df= df.drop(columns="Not_Useful_Column")
# drop the unecessary column in the dataset
```

```python
df
# printing it without that column
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123./_")
# write the column name with df[column-name]
# strip removes all the unecessary characters, special char, num from the specific column etc.
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('NaN','')
df["Phone_Number"] = df["Phone_Number"].str.replace('/','')
df["Phone_Number"] = df["Phone_Number"].str.replace('-','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na','')
df["Phone_Number"] = df["Phone_Number"].str.replace('|','')
# It will remove all the other characters

```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('NaN','')
```

```python
df.rename(columns={"Paying Customer":'Paying_Customer'})
```

```python
df=df.fillna('')
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=='Y':
      df.drop(x,inplace=True)
```

```python
df=df.reset_index(drop=True)
```
